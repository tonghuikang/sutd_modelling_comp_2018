```python
#initialization of parameters
x, y = 1,250
fire_starting_point = (x,y)
landmark_matrix = []

###simulate the growth of fire. SImplify from Rothermal model - treat it like the true data
#can use the Italy dataset to verify
#ROS is a function of wind_speed, fire_spreading_rate, average_slope and firefighting_ablity
ROS = (base_rate**(r/23.56487+np.sin(slope)/14.28231))*r + 0.122533
ROS = ROS * 60/1600 #change to miles per hour
print ROS

total_list = []
counter = 0
# total_list.append((x,y,0))
xylist = [(x,y)]

```

```python
###loop from here
newxylist =[]
#spread of fire, one time step
for entry in xylist:
    newxylist.append([entry[0]+ROS*times_step[times_index]*random.random(), entry[1]+ROS*times_step[times_index]*random.random()])
    newxylist.append([entry[0]+ROS*times_step[times_index]*np.sin(wind_direction),entry[1]+ROS*times_step[times_index]*np.cos(wind_direction)])
for entry in newxylist:
    total_list.append(entry)
xylist = newxylist    


# print total_list
nplist = np.array(total_list)
# print nplist


fig = plt.figure(figsize=(12,12))

ax=plt.subplot(aspect='equal')
# circle1 = plt.Circle((40, 350), 200, color='r', clip_on=True)
# ax.add_artist(circle1)

plt.scatter(nplist[:,0],nplist[:,1],s=10, color='r')


# draw nodes
nx.draw_networkx_nodes(G, pos, with_labels = True, node_color=node_color, nodelist=nodelist, node_size=100)

# overwrite
nx.draw_networkx_nodes(G, pos, with_labels = True, 
                        node_color='g', nodelist=colored_list[:times_index], 
                        node_size=300)
print colored_list[:times_index]
    
nx.draw_networkx_nodes(G, pos, with_labels = True, 
                        node_color='g', nodelist=colored_list[times_index-1:times_index], 
                        node_size=1900)
print colored_list[times_index-1:times_index]

# draw edges
nx.draw_networkx_edges(G, pos)
edge_labels = nx.get_edge_attributes(G,'r')

# draw edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_labels)

plt.axis('off')
times_index = times_index + 1
```
